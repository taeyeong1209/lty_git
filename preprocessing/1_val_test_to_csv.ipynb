### 1-2.분 단위 이하로 데이터가 존재하는 경우
- mAcc, mGps, wHr <br>
분 단위로 만들어주고 특성값에 max, min, last, mean, median을 구하고 칼럼으로 추가해서 저장 <br>
각 항목별로 반복 수행 후 저장

```python
def process_gzip_files(folder_path, file_identifier):
    """
    주어진 폴더 안의 gzip 파일들을 처리하여 데이터프레임을 생성하는 함수

    :param folder_path: gzip 파일들이 있는 폴더 경로
    :param file_identifier: 처리할 파일의 이름에 포함되어야 하는 식별자
    :return: 생성된 데이터프레임
    """
    # 빈 데이터프레임 생성
    result_df = pd.DataFrame()

    # 주어진 폴더 내의 모든 파일에 대해 반복
    for file_name in os.listdir(folder_path):
        if file_name.endswith('.gzip') and file_identifier in file_name:
            file_path = os.path.join(folder_path, file_name)

            # Parquet 파일을 데이터프레임으로 읽어오기
            df = pd.read_parquet(file_path)

            # timestamp를 분 단위로 변경
            df['timestamp'] = df['timestamp'].dt.floor('min')

            # 시간 구간별로 집계하여 통계량 계산
            agg_df = df.groupby('timestamp').agg({
                **{col: ['last', 'mean', 'max', 'min', 'median'] for col in df.columns if col not in ['subject_id', 'timestamp']}
            })

            # MultiIndex 칼럼을 flat하게 만들기
            agg_df.columns = ['_'.join(col).strip() for col in agg_df.columns.values]

            # timestamp를 칼럼으로 이동
            agg_df.reset_index(inplace=True)

            # subject_id를 timestamp의 마지막 값으로 설정
            subject_id = df.groupby('timestamp')['subject_id'].last()
            agg_df = pd.merge(subject_id, agg_df, on='timestamp',how='left')

            # 최종 결과에 현재 파일의 데이터프레임 추가
            result_df = pd.concat([result_df, agg_df])

    return result_df
```

```python
user1 = process_gzip_files(challenge2024_dataset_path2,'ch2024_test__m_acc_part_5.parquet.gzip')
```

```python
user2 = process_gzip_files(challenge2024_dataset_path2,'ch2024_test__m_acc_part_6.parquet.gzip')
```

```python
user3 = process_gzip_files(challenge2024_dataset_path2,'ch2024_test__m_acc_part_7.parquet.gzip')
```

```python
user4 = process_gzip_files(challenge2024_dataset_path2,'ch2024_test__m_acc_part_8.parquet.gzip')
```

```python
final_df = pd.concat([user1,user2,user3,user4],axis=0)
```
