```python
import os
import pandas as pd
import glob
import matplotlib.pyplot as plt
```

```python
# 데이터 읽기 및 NaN 값 처리
val = pd.read_csv('val_24hr_median_0628.csv', low_memory=False)
val_label = pd.read_csv('val_label.csv', low_memory=False)

# 데이터 읽기 및 NaN 값 처리
test = pd.read_csv('test_24hr_median_0628.csv', low_memory=False)

test
```

```python
val.drop(columns=['level_2'],inplace=True)
test.drop(columns=['level_2'],inplace=True)
```

```python
val_label_copy = val_label.copy()

val_label_copy
```

```python
# 열 이름 변경
val_label_copy.rename(columns={'subject_id': 'user'}, inplace=True)
val.rename(columns={'subject_id': 'user'}, inplace=True)

# label 데이터의 user 형식을 숫자로 변환하는 함수
def convert_user_to_number(user_str):
    if isinstance(user_str, str):
        return int(user_str.replace('user', ''))
    elif isinstance(user_str, float):
        return int(user_str)
    return user_str

val_label_copy['user'] = val_label_copy['user'].apply(convert_user_to_number)

val_label_copy['user']
```

```python
# 'datetime' 열을 datetime 타입으로 변환
val_label_copy['datetime'] = pd.to_datetime(val_label_copy['date'])
val['datetime'] = pd.to_datetime(val['datetime'])

# 인덱스 설정
val.set_index(['datetime', 'user'], inplace=True)
val_label_copy.set_index('datetime', inplace=True)

# 사용자별 최소 및 최대 시간 추출
user_times = val_label_copy.groupby('user').apply(lambda x: (x.index.min(), x.index.max()))

# 전체 시간대 인덱스 생성 및 데이터프레임 생성
all_dfs = []
for user, (start, end) in user_times.items():
    all_times = pd.date_range(start=start, end=end, freq='24H')
    user_df = pd.DataFrame(index=all_times, columns=['value'])
    user_df.index.name = 'datetime'
    user_df['user'] = user
    user_df.set_index('user', append=True, inplace=True)
    all_dfs.append(user_df)

all_df = pd.concat(all_dfs)

# 데이터프레임 병합
# 정렬을 확인합니다.
all_df = all_df.sort_index()
val = val.sort_index()

# 데이터프레임을 병합합니다.
full_df = all_df.combine_first(val)
full_df = full_df.reset_index()

# 필요한 칼럼에 따라 정렬

# 사용자별 그룹화 후 datetime 기준으로 정렬
full_df = full_df.sort_values(by=['user', 'datetime'])
full_df
```

```python
final = pd.merge(full_df, val_label_copy, on=['user', 'datetime'], how='left')
final = final.dropna(subset=['Q1'])

final
```

```python
# 결측치 비율 계산
missing_ratio = final.isnull().mean() * 100

# 결과 출력
# 결측치 비율을 데이터프레임으로 정리
missing_ratio_df = pd.DataFrame(missing_ratio, columns=['Missing Ratio (%)'])
missing_ratio_df
```

```python
from fancyimpute import IterativeImputer
from sklearn.tree import DecisionTreeRegressor
from sklearn.experimental import enable_iterative_imputer

# 사용자별 그룹화 후 트리 기반 MICE 보간 적용
def tree_based_mice_impute(group):
    # DecisionTreeRegressor를 사용하여 IterativeImputer 설정
    imputer = IterativeImputer(estimator=DecisionTreeRegressor(), random_state=0)
    imputed = imputer.fit_transform(group)
    return pd.DataFrame(imputed, index=group.index, columns=group.columns)

# 보간할 열 선택
columns_to_impute = final.columns.difference(['datetime'])

# 보간할 열만 선택하여 데이터프레임 생성
df_to_impute = final[columns_to_impute]

df_to_impute
```

```python
# 사용자별 그룹화하여 MICE 보간 적용
imputed_df = df_to_impute.groupby('user').apply(tree_based_mice_impute).reset_index(drop=True)
```
